# Exploring Grib Files

```python
import iris
import iris.plot as iplt
import iris.quickplot as qplt
import matplotlib.pyplot as plt
plt.interactive(True)
```

## Load data

```python
cubes = iris.load('download.grib')
```

```python
len(cubes)
```

```python
print(cubes)
```

```python
pressure = cubes[14]
print(pressure)
```

```python
print(pressure)
```

```python
print(pressure)
```

```python
hour_11 = iris.Constraint(time=lambda cell: cell.point.hour == 11 & cell.point.day == 2)
```

```python
cube_11 = cubes.extract(hour_11)
```

```python
str(cube_11[14].coord('time'))
```

```python
pressure = cube_11[14]
print(pressure)
```

```python
contour = qplt.contour(pressure)
plt.gca().coastlines()
fig = plt.gcf()
fig.set_figheight(10)
fig.set_figwidth(10)
plt.show()
```
